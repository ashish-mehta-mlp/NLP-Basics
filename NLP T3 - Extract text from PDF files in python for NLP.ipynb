```python
import PyPDF2 as pdf
```

```python
file = open("A:/NLP Tutorials/introduction-to-natural-language-processing.pdf", "rb") #rb stands for read binary
```

```python
file
```

```python
#creating an object of pdf file reader

pdf_reader = pdf.PdfFileReader(file) 
```

```python
pdf_reader
```

```python
# check if the file is encrypted

pdf_reader.getIsEncrypted()
```

```python
# Checking how many pages in the pdf

pdf_reader.getNumPages()
```

```python
# extracting information from any page. Lets say page 13

page13 = pdf_reader.getPage(14)
```

```python
page13

# we can see that the page has been obtained in the binary form. We have to convert it into text data
```

```python
page13.extractText()
```

```python
# List of sentences in pdf

pdf_reader.getPage(15).extractText().split()
```

```python
page14 = pdf_reader.getPage(15)
```

### Append Write or Merge PDF's

1. Creating a new PDF file using the original PDF file

```python
# create a pdf writer object
pdf_writer = pdf.PdfFileWriter()
```

```python
# In the object, we add page13 that we extracted
pdf_writer.addPage(page13)

# to add another page in the pdf that we created
pdf_writer.addPage(page14)
```

```python
#creating a new pdf called pages
output = open("pages.pdf", "wb") #wb = write binary
```

```python
pdf_writer.write(output)
```

```python
output.close()
```

### Changing the sequence of pages in pdf

```python
# just change the sequence in which the pages are added
pdf_writer = pdf.PdfFileWriter()
pdf_writer.addPage(page14)
pdf_writer.addPage(page13)
output2 = open("pages2.pdf", "wb")
pdf_writer.write(output2)
output2.close()
```
